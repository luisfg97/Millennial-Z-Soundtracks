```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
canciones_populares = pd.read_csv('sql-csv/canciones_populares_artista.csv')
nivel_energia = pd.read_csv('sql-csv/CANCIONES_LVL_ENERGIA.csv')
generos_populares = pd.read_csv('sql-csv/NUMERO_GENEROS.csv')
apariciones = pd.read_csv('sql-csv/APARICIONES.csv')
popularidad = pd.read_csv('sql-csv/MEDIA_POPULARIDAD.csv')
pl_mañana = pd.read_csv('sql-csv/PLAYLIST_MAÑANA.csv')
pl_motivacional = pd.read_csv('sql-csv/PLAYLIST_MOTIVACIONAL.csv')
pl_noche = pd.read_csv('sql-csv/PLAYLIST_NOCHE.csv')
```

```python
plt.figure(figsize=(6,6))
plt.pie(nivel_energia['numero_canciones'], labels=nivel_energia['nivel_energia'], autopct="%1.1f%%", startangle=90, colors=["#ff922d", "#2ba5ca", "#f080ab"])
plt.title("Nivel de energia canciones")
plt.show()
```

```python
top_genres = ['pop', 'hip hop', 'rock']
generos_populares['genres'] = generos_populares['genre'].apply(lambda x: x if x in top_genres else 'otros')
df_agrupado = generos_populares.groupby('genres', as_index=False)['num_generos'].sum().sort_values('num_generos', ascending= False)

plt.figure(figsize=(12,6))
sns.barplot(data=df_agrupado, x="genres", y="num_generos")
plt.title("Número de canciones por género musical")
plt.xlabel("Géneros musicales")
plt.ylabel("Nº Canciones")
plt.show()
```
